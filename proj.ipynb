```
%matplotlib notebook
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
#from sklearn.neighbors import KNeighborsClassifier

np.set_printoptions(precision=2)

sleep = pd.read_excel('dataset.xlsx')
sleep = sleep.dropna()
feature_names_sleep = ['Avg_Work_Hrs_Day', 'BreakTime']
X_sleep = sleep[feature_names_sleep]
y_sleep = sleep['ID']
target_names_sleep = ['Sleep_loss']

#X_train, X_test, y_train, y_test = train_test_split(X_sleep, y_sleep, random_state = 0)

number_of_rows,number_of_cols = sleep.shape

from sklearn.model_selection import train_test_split

data_set_values=numpy.array(sleep)

train, test = train_test_split(data_set_values, test_size = 0.2,random_state=2,stratify=None)

X_train=train[:,:number_of_cols-1]
y_train=train[:,number_of_cols-1]

X_test=test[:,:number_of_cols-1]
y_test=test[:,number_of_cols-1]

print("Training data size: {}".format(train.shape))
print("Test data size: {}".format(test.shape))
```
